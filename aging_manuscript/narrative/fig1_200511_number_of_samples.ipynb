```python
p = inout.get_internal_path(
    'dynamic/tstoeger/181022_inclusive_any_detected/sample_meta.csv')
```

```python
df = pd.read_csv(p)
```

```python
tissues_to_consider = ['BAT', 'Blood', 'GutEP', 'MuscSat', 'Adrenal', 'Kidney',
       'Heart', 'Liver', 'Skin', 'Stomach', 'Brain', 'SI',
       'LI', 'Cerebellum', 'WAT', 'Esophagus', 'Lung'] 

```

```python
df = df[
    (df['recommend_to_discard_181022']==0) & 
    (df['tissue'].isin(tissues_to_consider)) &
    (df['pfu']==0) &
    (df['age'].isin([4, 9, 12, 18, 24]))
]
```

```python
renamer = {
        'Adrenal': 'Adrenal Gland', 
        'BAT': 'Brown adipose tissue', 
        'Blood': 'Blood', 
        'Brain': 'Frontal cortex', 
        'Esophagus': 'Esophagus', 
        'GutEP': 'Gut epithelium', 
        'Heart': 'Heart',
        'Kidney': 'Kidney', 
        'LI': 'Large intestine', 
        'Liver': 'Liver', 
        'Lung': 'Lung', 
        'MuscSat': 'Muscle', 
        'SI': 'Small intestine', 
        'Skin': 'Skin',
        'Stomach': 'Stomach', 
        'WAT': 'White adipose tissue', 
        'Cerebellum': 'Cerebellum'}
    
df['tissue'] = df['tissue'].replace(renamer)
```

```python
mice_per_condition = df.groupby(['tissue', 'age']).size().to_frame('mice').reset_index().pivot(
    index='tissue',
    columns='age',
    values='mice'
)
```
